```python
import numpy as np
from holopy.core.process import bg_correct
import holopy as hp
import matplotlib 
import matplotlib.pyplot as plt

an_image = matplotlib.image.imread("./h.jpeg")
plt.imshow(an_image)

rgb_weights = [0.2989, 0.5870, 0.1140]
gr = np.dot(an_image[...,:3], rgb_weights)
plt.imshow(gr, cmap=plt.get_cmap("gray"))

imagepath = '2image.jpg'
raw_holo = hp.load_image(gr, spacing = 0.0851)



bg = hp.load_image(imagepath, spacing = 0.0851)
holo = bg_correct(raw_holo, bg)
hp.show(holo)
```
